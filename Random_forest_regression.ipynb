```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
col_list = list (media.columns) #coverting column in list
for col_name in col_list:  # loop searcing in list
    if media[col_name].dtypes== 'object':  #wherever we find obeject data
        media[col_name]=le.fit_transform(media[col_name])  #apply the label encoding
```

```python
X=media.drop(['cost'],axis=1)
y=media['cost']
```

```python
X
```

```python
y
```

```python
#Recursive feature selection
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.model_selection import train_test_split,StratifiedKFold
```

```python
rfe=RFE(RandomForestRegressor(),n_features_to_select=10)
```

```python
rfe.fit(X,y)
```

```python
rfe.ranking_
```

```python
rfe.get_support()
```

```python
feature_selected=X.columns[rfe.get_support()]
feature_selected
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train=rfe.transform(X_train)
X_test=rfe.transform(X_test)
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,pred)
```
